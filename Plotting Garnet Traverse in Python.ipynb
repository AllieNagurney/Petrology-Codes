```python
#Import the relevant Python libraries and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

import os
os.getcwd()

#Read the excel file
data = pd.read_excel("Example Garnet.xlsx",sheet_name="Example Garnet")


#Define the columns in Excel / variables in Python
x = data.Distance
y1 = data.Xalm
y2 = data.Xprp
y3 = data.Xsps
y4 = data.Xgrs

#Plot the graph
fig, ax = plt.subplots() # Create a figure containing one axes
ax.scatter(x, y1, color = 'red', marker = 's', s = 80)
ax.scatter(x, y2, color = 'blue', marker = 'o', s = 80)
ax.scatter(x, y3, color = 'lawngreen', marker = 'd', s = 80)
ax.scatter(x, y4, color = 'gold', marker = '^', s = 80)
plt.xlabel('Distance Across Grain (µm)', fontsize=12)
plt.ylabel('X alm, prp, sps, grs', fontsize=12)
plt.title('Example Garnet',fontsize=12)
plt.show()
```
