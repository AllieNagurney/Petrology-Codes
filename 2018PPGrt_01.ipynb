```python
#importing all colors, equations, etc. 
%pylab inline
import numpy as np
import matplotlib.pyplot as plt


#loading the data, skipping the first 13 rows because they don't have data that will be graphed in them, importing all columns of data
def load_data(fname):
    data = loadtxt(fname, skiprows=13,usecols=(0,1,2,3,4,5))
    return data

#defining the filename, you can change it or not from the werami generic name
filename= './garnet_comp.tab'
data = load_data(filename)

#Amount of CaO,MnO,FeO,MgO in Garnet, make sure the variables you set agree with the order of variables in werami
#The reshape value is dependent on the resolution you output in werami. This must match the resolution in werami
temp = linspace(min(data[:,0]),max(data[:,0]),100)
tempC = temp - 273.0
press = linspace(min(data[:,1]),max(data[:,1]),100)
presskb = press / 1000
XCaO = (data[:,2].reshape(100,100))
XMnO = (data[:,3].reshape(100,100))
XFeO = (data[:,4].reshape(100,100))
XMgO = (data[:,5].reshape(100,100))

plt.figure()
#N decides the number of contour intervals, you can play with this
N = 10
CS = plt.contour(tempC,presskb,XCaO,N,cmap='jet')

plt.title("X CaO")
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Pressure (Kbar)")
plt.clabel(CS, inline=1, fontsize=10)
```

```python
#importing all colors, equations, etc. 
%pylab inline
import numpy as np
import matplotlib.pyplot as plt


#loading the data, skipping the first 13 rows because they don't have data that will be graphed in them, importing all columns of data
def load_data(fname):
    data = loadtxt(fname, skiprows=13,usecols=(0,1,2,3))
    return data

#defining the new file name
filename= './garnet_staurolite_volume.tab'
data = load_data(filename)

#Volume of Garnet and Staurolite
temp = linspace(min(data[:,0]),max(data[:,0]),100)
tempC = temp - 273.0
press = linspace(min(data[:,1]),max(data[:,1]),100)
presskb = press / 1000
Garnet = (data[:,2].reshape(100,100))
Staurolite = (data[:,3].reshape(100,100))

plt.figure()
N = 10
CS = plt.contour(tempC,presskb,Garnet,N,cmap='jet')

plt.title("Garnet")
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Pressure (Kbar)")
plt.clabel(CS, inline=1, fontsize=10)
```

```python
plt.figure()
N = 4
CS = plt.contour(tempC,presskb,Staurolite,N,cmap='jet')

plt.title("Staurolite")
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Pressure (Kbar)")
plt.clabel(CS, inline=1, fontsize=10)
```
